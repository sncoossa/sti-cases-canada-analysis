```python
from pathlib import Path

import numpy as np
import pandas as pd

pd.set_option("display.max_columns", 100)
```

```python
path = Path.cwd().joinpath()
```

```python
# Reading the Gonorrhea 2021 and Chlamydia 2021 Datasets for Province/Territory, Cases, and Rates
```

```python
gon2021_df = pd.read_csv(
    path / "2021-gonorrhea.csv", 
)
print("gon2021_df.shape", gon2021_df.shape)
gon2021_df.head()
```

```python
ch2021_df = pd.read_csv(
    path / "2021-chlamydia.csv",
)
print("ch2021_df.shape", ch2021_df.shape)
ch2021_df.head()
```

```python
# First set of data visualizations for the 2021 Datasets
```

```python
import matplotlib
import matplotlib.pyplot as plt

pt = gon2021_df['jurisdiction']

gon2021_df['cases'] = gon2021_df['cases'].astype(float)
cases = gon2021_df['cases']

plt.bar(pt, cases)
plt.title('2021 Gonorrhea Province/Territory vs. Cases')
plt.xlabel('Province/Territory')
plt.ylabel('Cases')
plt.show()
```

```python
pt1 = ch2021_df['jurisdiction']

ch2021_df['cases'] = ch2021_df['cases'].astype(float)
cases1 = ch2021_df['cases']

plt.bar(pt1, cases1)
plt.title('2021 Chlamydia Province/Territory vs. Cases')
plt.xlabel('Province/Territory')
plt.ylabel('Cases')
plt.show()
```

```python
pt2 = gon2021_df['jurisdiction']

gon2021_df['rate'] = gon2021_df['rate'].astype(float)
rate_gon = gon2021_df['rate']

plt.bar(pt2, rate_gon)
plt.title('2021 Gonorrhea Province/Territory vs. Rate')
plt.xlabel('Province/Territory')
plt.ylabel('Rate (per 100,000 population)')
plt.show()
```

```python
pt3 = ch2021_df['jurisdiction']

ch2021_df['rate'] = ch2021_df['rate'].astype(float)
rate_ch = ch2021_df['rate']

plt.bar(pt3, rate_ch)
plt.title('2021 Chlamydia Province/Territory vs. Rate')
plt.xlabel('Province/Territory')
plt.ylabel('Rate (per 100,000 population)')
plt.show()
```

```python
# Reading the 2011-2020 Chlamydia and Gonorrhea Datasets
```

```python
ch1_df = pd.read_csv(
    path / "2011-2020_chlamydia.csv", 
)
print("ch1_df.shape", ch1_df.shape)
ch1_df.head()
```

```python
gon1_df = pd.read_csv(
    path / "2011-2020_gonorrhea.csv", 
)
print("gon1_df.shape", gon1_df.shape)
gon1_df.head()
```

```python
year = ch1_df['Year']

ch1_df['Total'] = ch1_df['Total'].astype(float)
total = ch1_df['Total']

plt.bar(year, total)
plt.title('2011-2020 Chlamydia Total Rates in Canada')
plt.xlabel('Year')
plt.ylabel('Rate in Canada (per 100,000 cases)')
plt.show()
```

```python
year1 = gon1_df['Year']

gon1_df['Total'] = gon1_df['Total'].astype(float)
total1 = gon1_df['Total']

plt.bar(year1, total1)
plt.title('2011-2020 Gonorrhea Total Rates in Canada')
plt.xlabel('Year')
plt.ylabel('Rate in Canada (per 100,000 cases)')
plt.show()
```

```python
# Visualizing by year + gender, 
```

```python
ch_year = ch1_df['Year']
ch_gender = {
    'Female': (ch1_df['Female']),
    'Male': (ch1_df['Male']),
}

x = np.arange(len(ch_year))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in ch_gender.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Cases per 100,000 people')
ax.set_xlabel('Year')
ax.set_title('Gender Differences in Chlamydia Rates, 2011-2020')
ax.set_xticks(x + width, ch_year)
ax.legend(loc='upper left', ncols=10)
ax.set_ylim(0, 500)

plt.show()
```
